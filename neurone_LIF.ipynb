```python
### Exercice d'apprentissage Neuro.comp ###

import numpy as np 
import matplotlib.pyplot as plt 

# Paramètres de mon neurone 

V_repos = -70.0
V_seuil = -55.0
V_reset = -70.0
tau = 10.0
R = 10.0

# Paramètres de simulation 

dt = 0.1
T = 100.0
I = 2.0

```

```python
# vecteur de temps 

t = np.arange(0, T, dt)

# vectuer potentiel

V = np.zeros(len(t))

V[0] = V_repos

print("Nombre de pas de temps :", len(t))
print("Premier potentiel :", V[0])
```

```python
# Liste pour spikes 

spikes = []

# première simulation

for i in range(1, len(t)):
    dV = (-(V[i-1] - V_repos) + R * I) / tau
    V[i] = V[i-1] + dV * dt 
    if V[i] >= V_seuil:
        V[i] = V_reset
        spikes.append(t[i])

print(f"Nombre de spikes : {len(spikes)}")
print(f"Premier spike à : {spikes[0]:.1f} ms")
```

```python
plt.figure(figsize=(12, 4))

plt.plot(t, V, color="blue", linewidth=1)

# Spikes avec points rouges 

for spike in spikes:
    plt.axvline(x=spike, color="red", linestyle="--", alpha=0.5)

# Lignes de référence :

plt.axhline(y=V_seuil, color="orange", linestyle=":", label="Seuil (-55 mV)")
plt.axhline(y=V_repos, color="green", linestyle=":", label="Repos (-70 mV)")

## Etiquettes 

plt.xlabel("Temps (ms)")
plt.ylabel("Potentiel membranaire (mV)")
plt.title("Simulation d'un neurone LIF")
plt.legend()
plt.tight_layout()
plt.show()

```
